```python
def game_play(name, cnt):
    print(name*cnt)

def wrong_play(name, cnt):
    if name not in players:
        print(name, '없는 이름입니다')
        return -1
    if cnt < 1 or cnt >4:
        print('1~4의 숫자만 가능합니다! 입력한 숫자',cnt)
        return -1
```

```python
for i in range(5):
    next_player, count = input('이름'), int(input('횟수'))
    check = wrong_play(next_player, count)
    if check == -1:
        break
    else:
        game_play(next_player, count)
```

```python
import qrcode
from googlesearch import search

def generate_qrcode(data, filename):
    qr = qrcode.make(data)
    qr.save(filename)

# 키워드와 가져올 URL 개수 설정
keyword = 'site'
num_results = 10

# URL 가져오기
urls = list(search(keyword, num_results=num_results))
# 'qrcodes' 폴더가 없다면 생성
output_folder = 'qrcodes'
os.makedirs(output_folder, exist_ok=True)

# 각 URL에서 도메인의 두 번째 부분을 QR 코드로 저장
for i, url in enumerate(urls, start=1):
    domain_second_part = url.split('.')[1]
    filename = os.path.join(output_folder, f"qrcode_{i}.png")
    generate_qrcode(domain_second_part, filename)
    print(f"{filename} 파일이 생성되었습니다.")
```

```python
def reward_penalty(kind, points):
    #전역변수 명시
    global current_point
    print('현재 점수:', current_point)
    if kind =='상점':
        current_point += points
        print('상점 입력 후 점수:',current_point)
    elif kind =='벌점':
        current_point -= points
        print('벌점 입력 후 점수:',current_point)
    return current_point

current_point = 5
reward_penalty('상점',5)
print('최종 점수:', current_point)
```
