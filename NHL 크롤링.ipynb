```python
import pandas as pd

# data1
data1 = []
for line in lines[18:-1]:
    tmp = line.split(' ')
    data1.append([tmp[0]])

df1 = pd.DataFrame(data1, columns=['Column1'])

# data2
data2 = []
for line in lines[18:-1]:
    tmp = line.split(' ')
    cnt = len(tmp)
    data = tmp[1:cnt - 17 + 1]

    if len(data) != 0:
        data2.append([data[0:len(data) - 1], data[-1]])

df2 = pd.DataFrame(data2, columns=['Column2_Part1', 'Column2_Part2'])

# data3
data3 = []
for line in lines[18:]:
    tmp = line.split(' ')
    data3.append([tmp[x - 17] for x in range(1, 17)])

df3 = pd.DataFrame(data3, columns=[f'Column3_{x}' for x in range(1, 17)])

# Merge dataframes
result_df = pd.concat([df1, df2, df3], axis=1)

# Display the resulting dataframe
print(result_df)
```

```python
# 'Column2_Part1' 열을 삭제
result_df = result_df.drop('Column2_Part1', axis=1)

# Display the resulting dataframe
result_df
```

```python
# 마지막 행 삭제
result_df = result_df.drop(result_df.index[-1])

# Display the resulting dataframe
result_df
```

```python
# Rename columns
column_mapping = {'Column1': 'Season', 'Column2_Part2': 'Team', 'Column3_1': 'GP', 'Column3_2': 'G', 'Column3_3': 'A', 'Column3_4': 'P',
                  'Column3_5': '+/-', 'Column3_6': 'PIM', 'Column3_7': 'PPG', 'Column3_8': 'PPP', 'Column3_9': 'SHG', 'Column3_10': 'SHP',
                  'Column3_11': 'TOI/G', 'Column3_12': 'GWG', 'Column3_13': 'OTG', 'Column3_14': 'S', 'Column3_15': 'S%', 'Column3_16': 'FO%'}

result_df = result_df.rename(columns=column_mapping)

# Display the resulting dataframe with new column names
result_df
```

```python
import pandas as pd
import os

# 합칠 엑셀 파일들이 있는 디렉토리 경로
directory_path = 'C:/Users/USER/Downloads/ai 수업/1월 둘째주/새 폴더/새 폴더'

# 모든 엑셀 파일을 저장할 빈 리스트 생성
all_dfs = []

# 디렉토리 내의 모든 파일을 탐색
for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):  # 확장자가 xlsx인 파일만 처리
        file_path = os.path.join(directory_path, filename)

        # 엑셀 파일을 DataFrame으로 읽어와서 리스트에 추가
        df = pd.read_csv(file_path)
        all_dfs.append(df)

# 리스트에 있는 모든 DataFrame을 합치기
combined_df = pd.concat(all_dfs, ignore_index=True)

# 결과를 새로운 엑셀 파일로 저장
output_path = 'C:/Users/USER/Downloads/ai 수업/1월 둘째주/새 폴더/새 폴더/combine.csv'
combined_df.to_csv(output_path, index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv('C:/Users/USER/Downloads/ai 수업/1월 둘째주/새 폴더/새 폴더/PIT_Skaters.csv')

# 데이터 확인
display(df)

# 날짜와 온도 데이터 추출
date = df["Season"]
temp = df["P"]
season = df["Name"]

# 선 그래프 그리기
plt.figure(figsize=(10, 6))

# seaborn을 사용하여 선 그래프 그리기
sns.lineplot(x=date, y=temp, hue=season, marker="o")

# 각 데이터 포인트에 숫자 표시
for i, txt in enumerate(temp):
    plt.annotate(txt, (date[i], temp[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel("Player")
plt.ylabel("Points")
plt.title("최근 5년 간 공격수 포인트 수")
plt.legend(title="Season", bbox_to_anchor=(1, 1))

plt.show()
```
