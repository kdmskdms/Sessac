```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd

chrome_options = Options()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-gpu')  

chromeDriver = webdriver.Chrome(options=chrome_options)
chromeDriver.get("https://www.nhl.com/redwings/player/patrick-kane-8474141")

# 페이지 로딩 대기 (필요에 따라 적절한 대기 시간을 설정하세요)
chromeDriver.implicitly_wait(10)

# XPath를 사용하여 원하는 요소 찾기
target_elements = chromeDriver.find_elements(By.XPATH, '//*[@id="simple-tabpanel-0"]/div[2]/div')

# 찾은 요소가 있다면 첫 번째 요소의 텍스트 추출
if target_elements:
    target_element = target_elements[0]
    
    # 추출한 텍스트를 줄 단위로 나누어 리스트로 만들기
    lines = target_element.text.split('\n')
    print(lines)

# # 브라우저 종료
chromeDriver.quit()
```

```python
import pandas as pd

# data1
data1 = []
for line in lines[18:-1]:
    tmp = line.split(' ')
    data1.append([tmp[0]])

df1 = pd.DataFrame(data1, columns=['Column1'])

# data2
data2 = []
for line in lines[18:-1]:
    tmp = line.split(' ')
    cnt = len(tmp)
    data = tmp[1:cnt - 17 + 1]

    if len(data) != 0:
        data2.append([data[0:len(data) - 1], data[-1]])

df2 = pd.DataFrame(data2, columns=['Column2_Part1', 'Column2_Part2'])

# data3
data3 = []
for line in lines[18:]:
    tmp = line.split(' ')
    data3.append([tmp[x - 17] for x in range(1, 17)])

df3 = pd.DataFrame(data3, columns=[f'Column3_{x}' for x in range(1, 17)])

# Merge dataframes
result_df = pd.concat([df1, df2, df3], axis=1)

# Display the resulting dataframe
print(result_df)
```

```python
# 마지막 행 삭제
result_df = result_df.drop(result_df.index[-1])

# Display the resulting dataframe
result_df
```

```python
# Rename columns
column_mapping = {'Column1': 'Season', 'Column2_Part1': 'Team','Column2_Part2': 'Team2', 'Column3_1': 'GP', 'Column3_2': 'G', 'Column3_3': 'A', 'Column3_4': 'P',
                  'Column3_5': '+/-', 'Column3_6': 'PIM', 'Column3_7': 'PPG', 'Column3_8': 'PPP', 'Column3_9': 'SHG', 'Column3_10': 'SHP',
                  'Column3_11': 'TOI/G', 'Column3_12': 'GWG', 'Column3_13': 'OTG', 'Column3_14': 'S', 'Column3_15': 'S%', 'Column3_16': 'FO%'}

result_df = result_df.rename(columns=column_mapping)

# Display the resulting dataframe with new column names
result_df
```
